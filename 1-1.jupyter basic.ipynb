# 주피터 노트북 사용법
* Shift + Enter 키를 누르면 셀이 실행되고 다음 셀로 이동합니다.
* Enter 키를 누르면 다시 편집 상태로 돌아옵니다.
* ESC 키를 누르고
    * a 키를 누르면 위에 셀이 추가됩니다.
    * b 키를 누르면 아래에 셀이 추가됩니다.
    * dd 키를 누르면 셀이 삭제됩니다.
    * m 키를 누르면 문서셀로 변경됩니다.
    * y 키를 누르면 코드셀로 변경됩니다.
    * h 키를 누르면 단축키 목록을 볼 수 있습니다.
