#### Control Flow

```python
# for 반복문으로 리스트 원소 출력
for i in lang:
    print(i)
```

```python
# python 이외는 "기타"로 출력하는 코드
for i in lang:
    if i == "python":
        print("python")
    else:
        print("기타")
```

```python
# 특정 횟수만큼 반복하여 실행
for i in range(5):
    print(i)
```

```python
count = len(lang)

for i in range(count):
    print(lang[i])
```

```python
# 짝수일 때 파이썬, 홀수일 때 자바 출력
for i in range(1, 10):
    if i % 2 == 0:
        print("python")
    else:
        print("java")
```

```python
# 인덱스 번호와 값을 함께 출력 (enumerate)
for i, val in enumerate(lang):
    print(i, val)
```

```python
#### 문자열
```

```python
address = " 경기도 성남시 분당구 불정로 6 NAVER 그린팩토리 16층 "
address
```

```python
# 앞 뒤 공백제거
address = address.strip()
address
```

```python
# 문자열 길이
len(address)
```

```python
# 문자열 분리 - default값: 공백
address_list = address.split()
address_list
```

```python
# 슬라이싱
address[:2]
```

```python
# 특정문자 포함여부 확인
address.startswith("경기")
```

```python
"경기" in address
```

```python
#### 리스트
```

```python
gu = address_list[2]
gu
```

```python
street = address_list[3]
street
```

```python
# 마지막 원소 출력
address_list[-1]
```

```python
# 리스트로 분리된 문자열 연결 - " ".join()
" ".join(address_list)
```

```python
# 리스트 포함 여부 (모든 글자가 일치해야함)
"경기도" in address_list
```
