## 여러 파일을 한번에 로드하는 방법

```python
file_csv = glob("data/store/*.csv")
file_csv

# 한글 오류
# 타이핑 인코딩과 텍스트 인코딩 방식에 차이가 있어
# 다른 결과를 가져오기 때문에 소상*.csv를 입력해도 데이터가 안불러와질 수 있음
```

```python
pd.read_csv(file_csv[0])
```

```python
# file_list에 리스트 형태로 file_csv 파일 넣기
file_list = []

for file_csv_name in file_csv:
    print(file_csv_name)
    temp = pd.read_csv(file_csv_name, low_memory=False)
    file_list.append(temp)
```

```python
# 여러개(17개)의 파일을 하나의 데이터프레임으로 합치기
# concat : 합치기
df = pd.concat(file_list)
df.shape
```
